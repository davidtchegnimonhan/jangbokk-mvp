```python
from bitcoinrpc.authproxy import AuthServiceProxy
from pprint import pprint
import logging, csv, time
import codecs
import time
```

```python
liste = [["Numero du bloc", "Hash du bloc", "BTC devant etre crees", "Frais totaux", "BTC crees + frais totaux (demandes par le mineur)", "BTC vraiment crees", "Hash tx coinbase", "Hash prochain block"]]
    
    
for k in range(840000,842900):
    
    #ON CONNECTE PYTHON A SON NOEUD MAINNET
    logging.basicConfig()
    logging.getLogger("BitcoinRPC").setLevel(logging.NOTSET)
    rpc_user = "bitcoinnode"
    rpc_pass = "bitcoinnodetest"
    rpc_host = "127.0.0.1"
    rpc_client = AuthServiceProxy(f"http://{rpc_user}:{rpc_pass}@{rpc_host}:8332")
    
    #ON OBTIENT LE HASH DU BLOC K
    block_hash = rpc_client.getblockhash(k)
    
    #ON OBTIENT LE NB DE BTC QUI DEVRAIENT ETRE CREES, AINSI QUE LES FRAIS TOTAUX DU BLOC
    block_stats = rpc_client.getblockstats(block_hash)
    block_subsidy = block_stats['subsidy'] #MONTANT DE BITCOINS QUI DEVRAIENT ETRE CREES NORMALEMENT
    block_totalfee = block_stats['totalfee'] #FRAIS TOTAUX DU BLOC
    
    #ON OBTIENT LES HASH DES TX DU BLOC, EN PARTICULIER LA COINBASE & LE HASH DU BLOC QUI SUIT
    block = rpc_client.getblock(block_hash)
    block_coinbase = block['tx'][0] #COINBASE DU BLOCK
    block_nextblockhash = block['nextblockhash'] #HASH DU BLOC SUIVANT
    
    #ON DECODE LE RAW DE LA TX COINBASE, ET ON SOMME LA TOTALITE DES OUTPUTS ASSOCIES
    coinbase=rpc_client.getrawtransaction(block_coinbase)
    output_coinbase=rpc_client.decoderawtransaction(coinbase)['vout']
    total_output_coinbase = 0
    
    for j in range(len(output_coinbase)):
        total_output_coinbase = (total_output_coinbase + output_coinbase[j]['value'])
        
    total_output_coinbase = int(total_output_coinbase*100000000)
        
        
        
    #ON CREEE LA LISTE FINALE QU'ON VA CONCATENER POUR DONNER NOTRE FICHIER CSV FUTUR
    row = [k,block_hash,block_subsidy,block_totalfee,total_output_coinbase,total_output_coinbase-block_totalfee,block_coinbase,block_nextblockhash]
    liste.append(row)
    
    print(k)
```

```python
import csv

with open("excelfinal.csv","w") as csvfile:
    writer = csv.writer(csvfile)
    writer.writerows(liste)
    
#on importe sur l'ordi ce fichier sous format csv
```
